```python
def my_min(iterable):
    # create and initialise the min_val to positive infinity
    min_val = float('inf')  
    # using for loop to loop through each element
    for item in iterable:
        # the min_val is updatad any time the element is smaller or lesser than
        if item < min_val:
            min_val = item
    
    return min_val

def my_max(iterable):
    # create and initialise the min_val to negative infinity
    max_val = float('-inf')
    
    # using for loop to loop through each element
    for item in iterable:
        # the max_val is updatad any time the element is greeter than
        if item > max_val:
            max_val = item
    
    return max_val

# Here we test the Test minimum and maximum functions created 
numbers = [13, 3, 9, 2, 7,6]
print("Minimum:", my_min(numbers))  
print("Maximum:", my_max(numbers))  
# we will see an output of 2 and 13 
```

## Create function final function to execute Y = (xi- xminimum)/(xmaximum-xminimum) and apply it to the data, this function can scale the dataframe genedata from 0 to 1 in a normalise function that is written from scatch

```python
#  Function to normalize a DataFrame
def normalize_dataframe(dataframe):
    numeric_cols = dataframe.select_dtypes(include=['number']).columns
    min_vals = dataframe[numeric_cols].apply(my_min)
    max_vals = dataframe[numeric_cols].apply(my_max)
    normalized_df = (dataframe[numeric_cols] - min_vals) / (max_vals - min_vals)
    return pd.concat([normalized_df, dataframe.drop(columns=numeric_cols)], axis=1)

# Normalizing the entire DataFrame
normalized_data = normalize_dataframe(genedata)

print(normalized_data)
```
